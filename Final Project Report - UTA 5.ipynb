### RFormula

### Removed Incident Location Type and Incident Categorical Description

```python
# Define an RFormula 

formula = RFormula(formula="ARC_Decision~.-Incident_Date",
    featuresCol="features",
    labelCol="label")

# Apply (fit) this formula to CD_data to retrieve and map CD_data's variables to the formula.
# Save the fitted formula

fittedARC = formula.fit(all_cities)


# Apply (transform) the fitted R formula to transform 
# the features and label variables in CD_data

preparedARC = fittedARC.transform(all_cities)


train, test = preparedARC.randomSplit([0.7,0.3], seed = 100)
```

```python
from pyspark.mllib.evaluation import MulticlassMetrics

def ARCDecision_classifier_evaluator (predictions):
    predictionAndLabels =predictions.select("prediction", "label").rdd
    metrics = MulticlassMetrics(predictionAndLabels)
    print("Precision(class1):", round(metrics.precision(0.0),3))
    print("Recall(class1):", round(metrics.recall(0.0),3))
    print("F-Score(class1):", round(metrics.fMeasure(0.0, beta=1.0),3))
    print("Precision(class2):", round(metrics.precision(1.0),3))
    print("Recall(class2):", round(metrics.recall(1.0),3))
    print("F-Score(class2):", round(metrics.fMeasure(1.0, beta=1.0),3))
    print("Precision(class3):", round(metrics.precision(2.0),3))
    print("Recall(class3):", round(metrics.recall(2.0),3))
    print("F-Score(class3):", round(metrics.fMeasure(2.0, beta=1.0),3))   
    print("Precision(class4):", round(metrics.precision(3.0),3))
    print("Recall(class4):", round(metrics.recall(3.0),3))
    print("F-Score(class4):", round(metrics.fMeasure(3.0, beta=1.0),3))
    print("Precision(class5):", round(metrics.precision(4.0),3))
    print("Recall(class5):", round(metrics.recall(4.0),3))
    print("F-Score(class5):", round(metrics.fMeasure(4.0, beta=1.0),3))
    print("Precision(class6):", round(metrics.precision(5.0),3))
    print("Recall(class6):", round(metrics.recall(5.0),3))
    print("F-Score(class6):", round(metrics.fMeasure(5.0, beta=1.0),3))
    print("Precision(class7):", round(metrics.precision(6.0),3))
    print("Recall(class7):", round(metrics.recall(6.0),3))
    print("F-Score(class7):", round(metrics.fMeasure(6.0, beta=1.0),3))
    print("Accuracy:", round(metrics.accuracy,3))  
```

```python
df = spark.read.csv("Clean_Accidents_v4.csv", header = True, inferSchema="true" )
df.groupBy('ARC_Decision').count().show()
```

```python
# Define an RFormula 

formula = RFormula(formula="ARC_Decision~.",
    featuresCol="features",
    labelCol="label")

# Apply (fit) this formula to df to retrieve and map df's variables to the formula.
# Save the fitted formula

fittedARC = formula.fit(df)

# Apply (transform) the fitted R formula to transform 
# the features and label variables in df

preparedARC = fittedARC.transform(df)
train, test = preparedARC.randomSplit([0.7,0.3], seed = 100)
```

```python
from pyspark.mllib.evaluation import MulticlassMetrics

def ARCDecision_classifier_evaluator (predictions):
    predictionAndLabels =predictions.select("prediction", "label").rdd
    metrics = MulticlassMetrics(predictionAndLabels)
    print("Precision(RO):", round(metrics.precision(0.0),3))
    print("Recall(RO):", round(metrics.recall(0.0),3))
    print("F-Score(RO):", round(metrics.fMeasure(0.0, beta=1.0),3))
    
    print("Precision(AV):", round(metrics.precision(1.0),3))
    print("Recall(AV):", round(metrics.recall(1.0),3))
    print("F-Score(AV):", round(metrics.fMeasure(1.0, beta=1.0),3))
    
    print("Precision(UN):", round(metrics.precision(2.0),3))
    print("Recall(UN):", round(metrics.recall(2.0),3))
    print("F-Score(UN):", round(metrics.fMeasure(2.0, beta=1.0),3))
    
    print("Accuracy:", round(metrics.accuracy,3))
```

## (I) Logistic Regression 

```python
# set a baseline Logistic Regression classifier
lr = LogisticRegression(labelCol='label',featuresCol='features', maxIter=20)

# Train the model using the train set
lr_fitted = lr.fit(train)

#Apply the trained model on test set 
lr_test_results = lr_fitted.transform(test)

# LogisticRegression has another function - evaluate to train a model 
lr_evaluated = lr_fitted.evaluate(test)

# Result
ARCDecision_classifier_evaluator(lr_test_results)
```

## (II) Decision Tree 

```python
# set a baseline decision tree classifier

dt = DecisionTreeClassifier(labelCol='label',featuresCol='features')

# Train a baseline decision tree model using the train set

dt_fitted = dt.fit(train)

# We can review the importance of features, typically determined based on information gain
dt_test_results = dt_fitted.transform(test)
ARCDecision_classifier_evaluator(dt_test_results)
```

```python
# Train a decision tree model that is pruned after three levels and view the rules
dt_max = DecisionTreeClassifier(labelCol='label'
                                ,featuresCol='features'
                                ,maxDepth = 9)
dt_fitted_max = dt_max.fit(train)

# Apply the simpler model to the test set and generate classification metrics
dt_max_test_results = dt_fitted_max.transform(test)
ARCDecision_classifier_evaluator(dt_max_test_results)
```

## (III) Random Forest

```python
# Improve random forst model 
rfc = RandomForestClassifier(labelCol='label'
                             ,featuresCol='features'
                             ,numTrees = 1
                             ,maxDepth = 4)
fitted_rfc = rfc.fit(train)

rfc_test_results = fitted_rfc.transform(test)
ARCDecision_classifier_evaluator(rfc_test_results)
```
